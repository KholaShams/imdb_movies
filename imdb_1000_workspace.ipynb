```python
def repl(string):

 return string.replace("[","").replace("u","").replace("'","").replace("]","")

dataset_actors =movies ['actors_list'].apply(repl)
```

```python
dataset_actors.head()
```

```python
actors = []
for i in dataset_actors:
    actors.append([j.strip() 
    for j in i.split(", ")])
```

```python
for i in actors:
    print(i)
```

```python
dictionary = {}
```

```python
for actor in actors:
    for a in actor:
        if (a in dictionary):
            dictionary[a] +=1
            
        else:
            dictionary[a]=1
    
       
```
