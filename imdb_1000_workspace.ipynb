```python
import pandas as pd
```

```python
movies = pd.read_csv('imdb_1000.csv')
```

```python
movies= pd.DataFrame(movies)
movies
```

```python
movies.head()
```

```python
movies.tail()
```

```python
#datatye of each column
movies.dtypes
```

```python
#average duration of each movie
averageOfMovies = movies.groupby('title')['duration'].mean()
print(averageOfMovies.head())
```

```python
#new set for movies with duration more than 150
new_movies = movies.loc[movies['duration']>=150]
new_movies
```

```python
new_movies.head()
```

```python
new_movies.tail()
```

```python
#count of content_rating
rating_content = movies.groupby('content_rating').sum()
rating_content
```

```python
rating_content.hist()
```

```python
rating = movies.groupby('genre')['star_rating'].mean()
rating
```

```python
rating.hist()
```
